```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_columns=[
    'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod','Churn','customerID']
```

```python
label_encoder =LabelEncoder()
```

```python
for col in categorical_columns:
  tele[col]=label_encoder.fit_transform(tele[col])
```

```python
le = LabelEncoder()
tele["customerID_e"] = le.fit_transform(tele["customerID"])
tele["gender_e"] = le.fit_transform(tele["gender"])
tele["Partner_e"] = le.fit_transform(tele["Partner"])
tele["PhoneService_e"] = le.fit_transform(tele["PhoneService"])
tele["MultipleLines_e"] = le.fit_transform(tele["MultipleLines"])
tele["InternetService_e"] = le.fit_transform(tele["InternetService"])
tele["OnlineSecurity_e"] = le.fit_transform(tele["OnlineSecurity"])
tele["DeviceProtection_e"] = le.fit_transform(tele["DeviceProtection"])
tele["TechSupport_e"] = le.fit_transform(tele["TechSupport"])
tele["StreamingTV_e"] = le.fit_transform(tele["StreamingTV"])
tele["StreamingMovies_e"] = le.fit_transform(tele["StreamingMovies"])
tele["Contract_e"] = le.fit_transform(tele["Contract"])
tele["PaperlessBilling_e"] = le.fit_transform(tele["PaperlessBilling"])
tele["PaymentMethod_e"] = le.fit_transform(tele["PaymentMethod"])
tele["Churn_e"] = le.fit_transform(tele["Churn"])
tele
```
